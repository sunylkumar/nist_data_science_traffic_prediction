```python
import pandas as pd

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
log_df = pd.read_csv("/home/datascience/Downloads/cleaning_test_06_09.tsv",
                     sep='\t', error_bad_lines= False,
                     na_values=['-'])
```

```python
occupancyGrouped = log_df.groupby(pd.cut(log_df["speed"], np.arange(-100, 200, 20)))
a=(occupancyGrouped["flow"].mean()).plot(kind='bar')
a.set_ylabel("Average flow")
a.set_title("Speed vs average flow")
```

```python
occupancyGrouped = log_df.groupby(pd.cut(log_df["speed"], np.arange(-100, 200, 20)))
a= (occupancyGrouped["occupancy"].sum()/(occupancyGrouped["occupancy"].count()+1)).plot(kind='bar')
a.set_ylabel("Average occupancy")
a.set_title("Speed vs average occupancy")
```

```python
# We can infer that speed in the range (0,20) has the greatest average occupancy, 
# in general we can observe that average occupancy is higher for lower speeds.
```
