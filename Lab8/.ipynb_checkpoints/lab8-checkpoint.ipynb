```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
#traffic conditions

fig = plt.figure(figsize=(200,100))


fig = plt.figure()
themap = Basemap(projection='gall', 
              # with low resolution,
llcrnrlon=df_traffcond['longitude'].min()-0.2, llcrnrlat=df_traffcond['latitude'].min()-0.2,
    urcrnrlon=df_traffcond['longitude'].max()+0.2, urcrnrlat=df_traffcond['latitude'].max()+0.2    ,   
    resolution = 'h', 
                 lon_0=df_traffcond['longitude'].mean(),
                 lat_0=df_traffcond['latitude'].mean(),
              # And threshold 100000
              area_thresh = 100000.0)
themap.drawcoastlines()
themap.drawcountries()
themap.fillcontinents(color = 'gainsboro')
themap.drawmapboundary(fill_color='steelblue')
x, y = themap(df_traffcond['longitude'].values, df_traffcond['latitude'].values)
themap.plot(x, y, 
            'o',                    # marker shape
            color='Indigo',         # marker colour
            markersize=4            # marker size
            )

plt.show()
```

```python
#accidents
fig = plt.figure(figsize=(200,100))


fig = plt.figure()
themap = Basemap(projection='gall', 
              # with low resolution,
llcrnrlon=df_accidents['longitude'].min()-0.2, llcrnrlat=df_accidents['latitude'].min()-0.2,
    urcrnrlon=df_accidents['longitude'].max()+0.2, urcrnrlat=df_accidents['latitude'].max()+0.2    ,   
    resolution = 'h', 
              # And threshold 100000
              area_thresh = 100000.0)
themap.drawcoastlines()
themap.drawcountries()
themap.fillcontinents(color = 'gainsboro')
themap.drawmapboundary(fill_color='steelblue')
x, y = themap(df_accidents['longitude'].values, df_accidents['latitude'].values)
themap.plot(x, y, 
            'o',                    # marker shape
            color='Red',         # marker colour
            markersize=4            # marker size
            )

plt.show()
```

```python
#roadworks
fig = plt.figure(figsize=(200,100))


fig = plt.figure()
themap = Basemap(projection='gall', 
              # with low resolution,
llcrnrlon=df_roadwork['longitude'].min()-0.2, llcrnrlat=df_roadwork['latitude'].min()-0.2,
    urcrnrlon=df_roadwork['longitude'].max()+0.2, urcrnrlat=df_roadwork['latitude'].max()+0.2    ,   
    resolution = 'h', 
              # And threshold 100000
              area_thresh = 100000.0)
themap.drawcoastlines()
themap.drawcountries()
themap.fillcontinents(color = 'gainsboro')
themap.drawmapboundary(fill_color='steelblue')
x, y = themap(df_roadwork['longitude'].values, df_roadwork['latitude'].values)
themap.plot(x, y, 
            'o',                    # marker shape
            color='Yellow',         # marker colour
            markersize=4            # marker size
            )

plt.show()
```
