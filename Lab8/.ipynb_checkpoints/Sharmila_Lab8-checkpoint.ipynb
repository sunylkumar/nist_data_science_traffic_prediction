## Task #2: Plot Counts of Events by Year

```python
eventsDF = pd.read_csv('events_train_holdout.tsv', sep='\t', error_bad_lines=False)
```

```python
yearIndex=[""]
years=eventsDF['closed_tstamp'].values
years=years.tolist()

for i in range(len(years)):
  asd=(years[i])
  if isinstance(asd, str):
      if asd[0:2]=='20':        
          asd=(asd[0:4])            
          yearIndex.append(asd)

yearIndex=yearIndex[1:]

yearDF = pd.DataFrame({'year': yearIndex})
s= yearDF.groupby('year')
s.size().plot(kind='bar')
```

## Task 3: Estimate Distribution of measurements

```python
log_df = pd.read_csv('cleaning_test_06_09.tsv', sep='\t', error_bad_lines=False)
```

## Flow Distribution

```python
occupancyGrouped = log_df.groupby(pd.cut(log_df["flow"], np.arange(-50, 300, 20)))
(occupancyGrouped["flow"].count()/11417381).plot(kind='bar')
```

## Speed Distribution

```python
occupancyGrouped = log_df.groupby(pd.cut(log_df["speed"], np.arange(-100, 200, 20)))
(occupancyGrouped["speed"].count()/11417381).plot(kind='bar')
```

## Occupancy Distribution

```python
occupancyGrouped = log_df.groupby(pd.cut(log_df["occupancy"], np.arange(-10, 300, 20)))
(occupancyGrouped["occupancy"].count()/11417381).plot(kind='bar')
```

## Second Visulisation

```python
df_events = pd.read_csv("events_train_holdout.tsv", sep="\t",  error_bad_lines=False)
```

```python
df_responders = df_events[np.isfinite(df_events['number_of_responders'])]
```

```python
grouped= df_responders.groupby('event_type')
grouped['number_of_responders'].mean().plot(kind='bar')
```

```python
grouped= df_responders.groupby('event_type')
grouped['number_of_responders'].mean().sort_values(ascending=False).head(10)
```

### Conclusion: 
### The highest average number of responders is for Warning Advice that is 2.5455 and then  followed by Delas Status Cancellation which is 2.416

## Third Visulisation

```python
log_df = pd.read_csv("cleaning_test_06_09.tsv",
                     sep='\t', error_bad_lines= False,
                     na_values=['-'])
```

```python
occupancyGrouped = log_df.groupby(pd.cut(log_df["speed"], np.arange(-100, 200, 20)))
a=(occupancyGrouped["flow"].mean()).plot(kind='bar')
a.set_ylabel("Average flow")
a.set_title("Speed vs average flow")
```

```python
occupancyGrouped = log_df.groupby(pd.cut(log_df["speed"], np.arange(-100, 200, 20)))
a= (occupancyGrouped["occupancy"].sum()/(occupancyGrouped["occupancy"].count()+1)).plot(kind='bar')
a.set_ylabel("Average occupancy")
a.set_title("Speed vs average occupancy")
```

### Conclusion: 
### The highest average occupancy is for Speed which is in the range (0, 20) 
###  From the graph we can see that Lower speed ranges have higher average occupancy
